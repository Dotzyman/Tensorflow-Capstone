#LOAD DATASET

```python
fes = pd.read_csv('Foreign_Exchange_Rates.csv')
fes.head(5)
```

#DATA WRANGLING/CLEANING

```python
#Get dataset info
fes.info()
```

```python
print('The No of Columns is  - ' ,fes.shape[1])
print('The No of Rows is  - ' ,fes.shape[0])
```

```python
#Check for all missing values
fes.isnull().sum()
```

```python
fes.columns
```

```python
#drop unnamed column as it is relevant to our datatset 
fes = fes.drop(columns = ['Unnamed: 0'])
```

```python
fes.describe().T
```

```python
#replace ND values with NaN values
fes.replace('ND',np.nan, inplace = True)
```

```python
fes.describe().T
```

```python
fes.isna().sum()
```

```python
#rename Time serie as Date and convert to datetime
fes = fes.rename(columns = { 'Time Serie' : 'Date'})
fes['Date'] = pd.to_datetime(fes['Date'])
fes.head().style.set_properties(subset = ['Date'], **{'background-color' : 'dodgerblue'})
```

```python
fes.set_index('Date', inplace =True)
```

```python
#convert the rate to numerical data
fes = fes.apply(pd.to_numeric)
fes.info()
```

```python
fes = fes.bfill().ffill()
```

```python
fes.isna().sum()
```

```python
# plotting a correlation matrix as a heatmap;
plt.figure(figsize= (20, 8))
mask = np.tril(np.ones_like(fes.corr(), dtype= bool))
sns.heatmap(fes.corr(), cmap='BrBG', mask=mask, vmin=-1, vmax=1, fmt = '.3f', annot=True)
plt.show()
```

```python
fes.plot(figsize = (10, 5))

plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.title('Exchange Rates of Diff Currencies')

plt.show()
```
